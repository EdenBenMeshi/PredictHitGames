```python
def calculate_pages_amount(genre):
    html = open_url(genre,1);
    tables = html.find('div', id='generalBody').find('tr').find('th').text
    results = ''.join(filter(str.isdigit, tables))
    amount = round(int(results)/RESULTS+.5)
    print(f"the amount of results is {results} and the amount of pages is")
    print(amount)
    return amount
```

The function will crawls the HTML page and deliver the data to Lists.
           
**Parameters:**
    gets:
        html page, and crawls the content of the HTML page to lists 

```python
def extract_items(html_page):
    table = html_page.find('div', id='generalBody').find_all('tr')
    #Delete the first three tables and the last table 
    del table[:3] 
    del table[-1]
    for game in table:
        row = game.find_all('td')
        Pos.append(row[0].text.strip())
        Game.append(row[2].text.strip())
        Console.append(row[3].find('img')['alt'].strip())
        Publisher.append(row[4].text.strip())
        Developer.append(row[5].text.strip())
        VGChartz_Score.append(row[6].text.strip())
        Critic_Score.append(row[7].text.strip())
        User_Score.append(row[8].text.strip())
        Total_Shipped.append(row[9].text.strip())
        Total_Sales.append(row[10].text.strip())
        NA_Sales.append(row[11].text.strip())
        PAL_Sales.append(row[12].text.strip())
        Japan_Sales.append(row[13].text.strip())
        Other_Sales.append(row[14].text.strip())
        Release_Date.append(row[15].text.strip())
        Last_Update.append(row[16].text.strip())
        Genre.append(genre.strip())
```

***Main code:***

    The code performs the process of crawling by using the functions we 
    built earlier.
    
    process: loop on all the genre and calculates the number of pages for 
    each specific genre. 
    After running the whole genre, loads and extracts the information from 
    any genre, 
    Finally put the information into dataframe and open the data 
    thtough excel/csv file.
        

```python
#
for genre in genres_list:
    pages = calculate_pages_amount(genre)
    for page in range(1,pages+1):
        html = open_url(genre,page)
        extract_items(html)
df = pd.DataFrame({ 'Game': Game,
                    'Pos':Pos,
                    'Genre': Genre,
                    'Console': Console,
                    'Publisher' : Publisher,
                    'Developer': Developer,
                    'VGChartz Score': VGChartz_Score,
                    'Critic Score': Critic_Score,
                    'User Score':User_Score,
                    'Total Shipped': Total_Shipped,
                    'Total Sales': Total_Sales,
                    'NA Sales' : NA_Sales,
                    'PAL Sales': PAL_Sales,
                    'Japan Sales': Japan_Sales,
                    'Other_Sales': Other_Sales,
                    'Release_Date': Release_Date,
                    'Last_Update' : Last_Update})
df.to_excel("games.xlsx")
df.to_csv("games.csv")
print("Crawling finish successfully")
```
